```python
df_process=pd.read_csv("C:/Users/nsh65/Documents/POSCO_Ai-BigData_academy/Week4_BigData_Project/Data/MFG_PROCESS.csv",engine="python",encoding='EUC-KR')
df_process.head()
```

```python
df_machine=pd.read_csv("C:/Users/nsh65/Documents/POSCO_Ai-BigData_academy/Week4_BigData_Project/Data/MFG_MACHINE.csv",engine="python",encoding='EUC-KR')
df_machine.head()
```

```python
df_machine=df_machine.drop('DATE_1BLACK',axis=1)
# machine 날짜 데이터 불필요로 제거 .
```

```python
df_machine
```

```python
df_tat=pd.read_csv("C:/Users/nsh65/Documents/POSCO_Ai-BigData_academy/Week4_BigData_Project/Data/MFG_TAT.csv",engine="python",encoding='EUC-KR')
df_tat.head()
```

```python
df = pd.merge(df_machine, df_tat, on=['JUDGE','LOT_ID','PANEL_ID','TIME_1BLACK','TIME_2AG','TIME_3BUS','TIME_4DIELEC','TIME_5MGO','TIME_6PHOS','TIME_7PASTE','TIME_8AGING']) # 교집합 값 inner
df_result = pd.merge(df, df_process, on=['JUDGE','LOT_ID','PANEL_ID','TIME_1BLACK','TIME_2AG','TIME_3BUS','TIME_4DIELEC','TIME_5MGO','TIME_6PHOS','TIME_7PASTE','TIME_8AGING']) # 교집합 값 inner
```

```python
df_result
```

##  데이터 결측값 확인

```python
obj = df_result.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```

```python
df_result = df_result.interpolate(method='linear', limit_direction='forward')
# 선형보간법을 이용한 결측값 처리
```
